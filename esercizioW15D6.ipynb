```python
import os
beginner_ds = os.listdir("/Users/danielamuto/Documents/EPICODE/PYTHON EPICODE/beginner_datasets")
```

```python
null_ds = []
not_null_ds = []

for ds in beginner_ds:
    count = 0
    if ".csv" in ds:
        df = pd.read_csv("{/Users/danielamuto/Documents/EPICODE/PYTHON EPICODE/beginner_datasets}{ds}")
        for colonna in df.columns:
            if df[colonna].isnull().sum() > 0:
                count += 1
        if count > 0:
            null_ds.append(ds)
        else: 
            not_null_ds.append(ds)

print("DS con valori nulli: ", null_ds, '\n', sep='')
print("DS con valori non nulli: ", not_null_ds, sep='')
```
