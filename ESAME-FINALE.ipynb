```python
# Si chiede di trovare, per ogni continente: 
#  a. il numero di casi fin dall'inizio della pandemia 
#   b. la percentuale rispetto al totale mondiale del numero di casi

casi_totali_continente = covid.groupby('continent')['total_cases'].max()
print("Il numero totale di casi dall'inizio della pandemia per continente è: \n", casi_totali_continente)

casi_totali_mondiali = covid['total_cases'].max()
percentuale_continente = (casi_totali_continente / casi_totali_mondiali) * 100
print("La percentuale di casi per continente rispetto al totale mondiale del numero di casi è: \n", percentuale_continente)
```

```python
# Selezionare i dati relativi all'Italia nel 2022 e mostrare con dei grafici adeguati: 
#  a. l'evoluzione del casi totali dall'inizio alla fine dell'anno 
#   b. il numero di nuovi casi rispetto alla data 
#    k. poiché i nuovi casi vengono registrati settimanalmente, filtrare via i dati che non hanno misurazioni

italia_2022 = covid[(covid['location'] == 'Italy') & (covid['date'].dt.year == 2022)]
print(italia_2022)
```

```python
italia_2022 = italia_2022.dropna(subset= ['new_cases'])
print(italia_2022[['date', 'new_cases']].isnull().sum())
```

```python
plt.figure(figsize=(12, 6))
plt.plot(italia_2022['date'], italia_2022['total_cases'], color='blue', label='Casi Totali')
plt.title('Evoluzione dei Casi Totali in Italia nel 2022')
plt.xlabel('Data')
plt.ylabel('Casi Totali')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(italia_2022['date'], italia_2022['new_cases'], color='red', label='Nuovi Casi')
plt.title('Evoluzione dei Nuovi Casi Settimanali in Italia nel 2022')
plt.xlabel('Data')
plt.ylabel('Nuovi Casi')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Riguardo le nazioni di Italia, Germania e Francia mostrare in un boxplot la differenza tra queste nazioni riguardo 
# il numero di pazienti in terapia intensiva (Intensive Care Unit, ICU) da maggio 2022 (incluso) ad aprile 2023 (incluso)

pazienti = covid[(covid['location'].isin(['Italy', 'Germany', 'France'])) & (covid['date'] >= '2022-05-01') & (covid['date'] <= '2023-04-30')]
print(pazienti[['location', 'date', 'icu_patients']].head())
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=pazienti, x='location', y='icu_patients', palette='Set2')
plt.title('Distribuzione dei Pazienti in Terapia Intensiva (ICU) per Paese (Maggio 2022 - Aprile 2023)')
plt.xlabel('Paese')
plt.ylabel('Pazienti in Terapia Intensiva (ICU)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# commento grafico: 
#  Il grafico mostra appunto la distribuzione dei pazienti in terapia intensiva nel periodo tra maggio 2022 ed aprile 2023
#  per tre paesi europei (Italia, Germania e Francia). Possiamo notare che la distribuzione dei pazienti in Italia è molto
#  più concentrata rispetto agli altri due paesi, suggerendoci quindi che il numero di pazienti in terapia intensiva tendeva
#  ad essere più stabile. In Germania il numero di pazienti ha mostrato fluttuazioni più pronunciate. Anche la Francia, 
#  infine, mostra una certa variabilità, con un numero significativo di pazienti in terapia intensiva in alcuni periodi.
```

```python
# Riguardo le nazioni di Italia, Germania, Francia e Spagna in tutto il 2023: 
#  a. mostrare la somma dei pazienti ospitalizzati per ognuna 
#  b. se ci sono dati nulli, con un breve commento scrivere se può essere possibile gestirli tramite sostituzione o meno

ospedalizzati = covid[(covid['location'].isin(['Italy', 'Germany', 'France', 'Spain'])) & (covid['date'].dt.year == 2023)]
print(ospedalizzati[['location', 'date', 'hosp_patients']].head())
```

```python
ospedalizzati_per_nazione = ospedalizzati.groupby('location')['hosp_patients'].sum()
print("La somma del pazienti ospedalizzati per nazione è: \n", ospedalizzati_per_nazione)
```

```python
dati_mancanti = ospedalizzati['hosp_patients'].isnull().sum()

if dati_mancanti > 0:
    print(f"Sono presenti {dati_mancanti} valori nulli nei pazienti ospedalizzati.")
    print("I valori nulli possono essere gestiti tramite sostituzione con la mediana o la media, se non sono troppi. \nTuttavia, se i valori nulli sono troppi, potrebbero indicare un problema nei dati.")
else:
    print("Non sono presenti valori nulli nei pazienti ospedalizzati.")
```
