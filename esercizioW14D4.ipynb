```python
diabetes.shape
```

```python
diabetes.head(5)
```

```python
list(diabetes.columns)
```

```python
diabetes.describe()
```

```python
a = (diabetes['Age (years)']<20)
b = (diabetes['Age (years)']>=20) & (diabetes['Age (years)']<30)
c = (diabetes['Age (years)']>=30) & (diabetes['Age (years)']<40)
d = (diabetes['Age (years)']>=40) & (diabetes['Age (years)']<50)
e = (diabetes['Age (years)']>=50)
```

```python
p1 = (diabetes.loc[a, 'Diastolic blood pressure (mm Hg)']).mean()
p2 = (diabetes.loc[b, 'Diastolic blood pressure (mm Hg)']).mean()
p3 = (diabetes.loc[c, 'Diastolic blood pressure (mm Hg)']).mean()
p4 = (diabetes.loc[d, 'Diastolic blood pressure (mm Hg)']).mean()
p5 = (diabetes.loc[e, 'Diastolic blood pressure (mm Hg)']).mean()

print(f"La pressione sanguigna media per fasce di età è: \n<20 {p1} \n20-30 {p2} \n30-40 {p3} \n40-50 {p4} \n>50 {p5}")
```

```python
media_pressione_eta = diabetes.groupby('Age (years)')['Diastolic blood pressure (mm Hg)'].mean()
```

```python
pressione_eta = [media_pressione_eta, 'Diastolic blood pressure (mm Hg)']
print(pressione_eta)
```
